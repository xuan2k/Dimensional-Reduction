```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.manifold import TSNE
from sklearn.manifold import MDS
from sklearn import datasets, decomposition

# Generate example dataset with three classes in 2D
X, y = make_blobs(n_samples=300, centers=3, n_features=2, random_state=42)

# Visualize the original dataset
print(X.shape)
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='viridis')
plt.title("Original Dataset")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.savefig("original_dataset.png")
plt.show()

# Perform t-SNE for dimensionality reduction into 1D
tsne = TSNE(n_components=2, random_state=42, n_iter=10000)
X_tsne = tsne.fit_transform(X)

print(X_tsne.shape)
# Visualize the t-SNE result
plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=y, cmap='viridis')
# plt.scatter(X_tsne[:, 0], np.zeros_like(X_tsne), c=y, cmap='viridis')
plt.title("t-SNE Visualization")
plt.xlabel("t-SNE Component")
plt.ylabel("")
plt.savefig("tsne_result.png")
plt.show()

pca = decomposition.PCA(n_components=2)
X_pca = pca.fit_transform(X)

# Visualize the t-SNE result
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y, cmap='viridis')
# plt.scatter(X_tsne[:, 0], np.zeros_like(X_tsne), c=y, cmap='viridis')
plt.title("PCA Visualization")
plt.xlabel("PCA Component")
plt.ylabel("")
plt.savefig("pca_result.png")
plt.show()
# Perform SNE by (MDS) for dimensionality reduction into 1D
sne = MDS(n_components=1, random_state=42)
X_sne = sne.fit_transform(X)

# Visualize the SNE result
plt.scatter(X_sne, np.zeros_like(X_sne), c=y, cmap='viridis')
plt.title("SNE Visualization")
plt.xlabel("SNE Component")
plt.ylabel("")
plt.savefig("sne_result.png")
plt.show()
```
